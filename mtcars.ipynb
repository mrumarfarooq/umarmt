```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
plt.figure(figsize=(16,9)) # Figure size
sns.scatterplot(x='wt', y='mpg', data=df, marker='o', color='indigo') 
plt.title('milesperhour') # Title
#plt.xticks(df.model.unique(), rotation=45) 
plt.xlabel('wt')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(16,9)) # Figure size
sns.histplot(x='model', y='mpg', data=df, color='indigo') 
plt.title('frequency distribution of mpg') # Title
#plt.xticks(df.model.unique(), rotation=45) 
plt.xlabel('model')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(16,9)) # Figure size
sns.barplot(x='model', y='am', data=df, color='indigo') 
plt.title('frequency distribution of am') # Title
#plt.xticks(df.model.unique(), rotation=45) 
plt.xlabel('model')
plt.ylabel('am')
plt.show()
```

```python
plt.figure(figsize=(16,9)) # Figure size
sns.boxplot(x='model', y='mpg', data=df, color='indigo') 
plt.title('frequency distribution of am') # Title
#plt.xticks(df.model.unique(), rotation=45) 
plt.xlabel('model')
plt.ylabel('mpg')
plt.show()
```
